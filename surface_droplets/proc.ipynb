```python
import numpy as np
import pickle 
from datetime import datetime
```

```python
def load_data(path):
    res = pickle.load(open(path, "rb"))
    Rs = np.array([i[1] for i in res])
    Ts = np.array([datetime.strptime(i[0].split("_")[-1].split(".")[0], "%H%M%S").timestamp() for i in res])
    Ts = Ts - Ts[0]
    return Ts, Rs
```

```python
for i in ["d2_1_b13_res.pickle", "d2_2_res.pickle", "d2_3_res.pickle", "d2_4_b15_res.pickle", "d2_5_b13_wet_res.pickle"]:
    Ts, Rs = load_data(i)
    plt.plot(Ts, np.power(Rs, 1), label=i)

plt.legend()
```
